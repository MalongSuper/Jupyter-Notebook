### Data Analytics: Lab 6

MUST USE pandas.pivot_table(df,…)

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('mtcars.csv')

df
```

1. Group by cyl, then apply aggregate functions (mean) on ‘mpg’

```python
table = np.round(pd.pivot_table(df, index='cyl', values='mpg', aggfunc=np.mean), 2)
table
```

2. Group by cyl, then apply aggregate functions (mean) on [‘mpg’, ‘hp’]

```python
table = np.round(pd.pivot_table(df, index='cyl', values=['mpg', 'hp'], aggfunc=np.mean), 2)
table
```

3. Group by [cyl, am], then apply aggregate functions (mean) on [‘mpg’, ‘hp’]

```python
table = np.round(pd.pivot_table(df, index=['cyl', 'am'], values=['mpg', 'hp'], aggfunc=np.mean), 2)
table
```

4. Group by [cyl], then apply aggregate functions (mean, std) on [‘mpg’, ‘hp’]

```python
table = np.round(pd.pivot_table(df, index='cyl', values=['mpg', 'hp'], aggfunc=[np.mean, np.std]), 2)
table
```

5. Group by [cyl, am], then apply aggregate functions (mean, std) on [‘mpg’, ‘hp’]

```python
table = np.round(pd.pivot_table(df, index=['cyl', 'am'], values=['mpg', 'hp'], aggfunc=[np.mean, np.std]), 2)
table
```

6. Group by [cyl, am], then apply aggregate functions (mean, median, std) on ‘mpg’ and
then apply aggregate functions (min, max) on ‘hp’.

Also, use fill_value = 0, margins = True, margins_name = “Total”

```python
table = np.round(pd.pivot_table(df, index=['cyl', 'am'], 
                                values=['mpg', 'hp'], 
                                aggfunc={'mpg': ['mean', 'median', 'std'], 
                                         'hp':['min', 'max']}, 
                                         fill_value=0, 
                                         margins=False, margins_name='Total'), 2) # margins=True raises an error


table
```

7. Group by [cyl], set pivot table column as [gear], then apply aggregate functions (mean,
std) on [‘mpg’, ‘hp’]. Also, use fill_value = 0, margins = True, margins_name = “Total”.

```python
table = np.round(pd.pivot_table(df, index=['cyl'], columns=['gear'], values=['mpg', 'hp'], aggfunc=[np.mean, np.std], 
                                fill_value=0, 
                                         margins=True, margins_name='Total'), 2)
table
```

8. Group by [cyl, am], set pivot table columns as [gear, carb], then apply aggregate functions (mean, std) on [‘mpg’, ‘hp’]. Also, use fill_value = 0, margins = True, margins_name = “Total”.

```python
table = np.round(pd.pivot_table(df, index=['cyl', 'am'], columns=['gear', 'carb'], values=['mpg', 'hp'], aggfunc=[np.mean, np.std], 
                                fill_value=0, margins=True, margins_name='Total'), 2)
table
```

9. Group by [cyl, am], set pivot table columns as [gear, carb], then apply aggregate
functions (lambda func on: sqrt(sum(x*x))) on [‘mpg’, ‘hp’]. Also, use fill_value = 0,
margins = True, margins_name = “Total”.

```python
table = np.round(pd.pivot_table(df, index=['cyl', 'am'], columns=['gear', 'carb'], values=['mpg', 'hp'], 
                                aggfunc=lambda x: np.sqrt(sum(x*x)), 
                                fill_value=0, margins=True, margins_name='Total'), 2)
table
```
